# Пример исследования

Например мы хотим посмотреть как итеративно будет вести себя точка при применении функции из `Pandora`

```python
import src.ComplexNumber as CN
import src.Pandora as Pandora
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = CN.ComplexNumber(0, 0)
trajectory = [[x.real, x.imaginary]]
for _ in range(5):
    x = Pandora.Function1.func(x)
    trajectory.append([x.real, x.imaginary])
trajectory = np.array(trajectory).T
display(trajectory)
```

```python
plt.figure()
plt.scatter(*trajectory[:, 0], label='start')
plt.scatter(*trajectory[:, -1], label='finish')
plt.plot(trajectory[0], trajectory[1])
plt.legend()
plt.xlabel("Re")
plt.ylabel("Im")
plt.show()
```
