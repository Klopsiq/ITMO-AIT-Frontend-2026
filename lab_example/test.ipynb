# Тесты/демонстрация работы класса ComplexNumber

```python
import src.ComplexNumber as CN
```

```python
# Блок проверки на возможные конструкторы класса + адекватное отображение 
assert str(CN.ComplexNumber(1, 2)) == "1 + i * 2"
assert str(CN.ComplexNumber(-2, 4)) == "-2 + i * 4"
assert str(CN.ComplexNumber(2, -5)) == "2 + i * -5"
```

```python
# Проверка базового функционала класса
a = CN.ComplexNumber(1, 2)
b = CN.ComplexNumber(3, 5)

assert (a + b) == CN.ComplexNumber(4, 7)
assert (a * b) == CN.ComplexNumber(-7, 11)
```
